### PROBLEM 1
**The Lotka-Volterra equations**

_Refer to Exercise 8.2 in the Newman text._ The Lotka-Volterra equations are a mathematical model of predator-prey interactions between biological species. Let two variables $x$ and $y$ be proportional to the size of the populations of two species, traditionally called "rabbits" (the prey) and "foxes" (the predators). You could think of $x$ and $y$ as being the population in thousands, say, so that $x = 2$ means there are 2000 rabbits.

In the Lotka-Volterra model the rabbits reproduce at a rate proportional to their population, but are eaten by the foxes at a rate proportional to both their own population and the population of foxes:

$$\frac{{\rm d}x}{{\rm d}t} = \alpha x - \beta xy,$$

where $\alpha$ and $\beta$ are constants. At the same time the foxes reproduce at a rate proportional to the rate at which they eat rabbits-- because they need food to grow and reproduce-- but also die of old age at a rate proportional to their own population:

$$\frac{{\rm d}y}{{\rm d}t}  = \gamma xy - \delta y,$$

where $\gamma$ and $\delta$ are also constants. *(25 pts.)*

1. Solve these equations using the fourth-order Runge-Kutta method for the case $\alpha = 1, \beta = \gamma = 0.5$, and $\delta = 2$, starting from the initial condition $x = y = 2$.

2. Make a graph showing both $x$ and $y$ as a function of time on the same axes from $t =0$ to $t = 30$. *(25 pts.)*

```python
# Number 1

import numpy as np

# We set the values for the given constants

alpha = 1
beta = 0.5
gamma = 0.5
delta = 2

# We encode both x' and y' into one Python function f(r). This makes it more convenient to program the Runge-Kutta method later on.

def f(r):
    
    # We will use the array r as a sort of "vector".
    
    x = r[0] # The first entry of r gives the initial population of rabbits.
    y = r[1] # The second entry of r gives the initial population of foxes.
    x_prime = alpha*x - beta*x*y # x_prime is the time derivative of x. This is given.
    y_prime = gamma*x*y - delta*y # y_prime is the time derivative of y. This is also given.
    f = np.array([x_prime, y_prime])
    return f

""""

Despite being combined into a single function, notice how x' and y' are still separate from each other, 
and we could work with one or the other as we please.

If you'd like, you may think of f(r) as a sort of vector function that assigns to each point x and y 
the vector with components [x_prime, y_prime].

I still prefer to think of them as 2 separate equations x' and y'.

"""

# We define our own function for the 4th order Runge Kutta method.

def runge_kutta_4(f, r, t_0, t, N):
    
    h = (t - t_0)/N # This is the size of each step.
    
    tpoints = np.arange(t_0, t, h) # This array gives the discrete time values for our approximation in steps of h.
    xpoints = np.array([]) # We will store the values of x (Rabbits) in this array.
    ypoints = np.array([]) # We will store the values of y (Foxes) in this array.
    
    for t in tpoints:
        
        # For each discrete time step, we add the value of x and y to our arrays.
        
        xpoints = np.append(xpoints, r[0])
        ypoints = np.append(ypoints, r[1])
        
        # Below is simply the formula for the 4th order Runge Kutta method.
        
        k1 = h * f(r)
        k2 = h * f(r + 0.5*k1)
        k3 = h * f(r + 0.5*k2)
        k4 = h * f(r + k3)
        
        r += (k1 + 2*k2 + 2*k3 + k4)/6
        
    return xpoints, ypoints, tpoints

# We define some more constants.

t_0 = 0 # This will be the initial time in our graph.
t = 30 # This will be the end time.
N = 1000 # This is the amount of steps from our initial time to our end time.
r_0 = np.array([2, 2], float) # These are the initial values of x and y, stored in an array.

results = runge_kutta_4(f, r_0, t_0, t, N) # We call our function for our given values.

# We subset the results to obtain seperate arrays.

x = results[0]
y = results[1]
t = results[2]
```

```python
# Number 2

import matplotlib.pyplot as plt

# We plot the population of rabbits and foxes on the same graph.

fig, ax = plt.subplots()
plot = ax.plot(t, x, label = 'Rabbits')
plot1 = ax.plot(t, y, color = 'red', label = 'Foxes')

# We label our axes.

ax.set_title("Lotka-Voltera Equations Solutions")
ax.set_xlabel("Time (Years)")
ax.set_ylabel("Population (Thousands)")

# We display our plot and add a legend.

ax.legend()

plt.show()

# We plot a phase diagram for the population of foxes and rabbits.

fig1, ax1 = plt.subplots()
plot2 = ax1.plot(x, y)

# We label our axes.

ax1.set_title("Phase Diagram for Population of Rabbits and Foxes")
ax1.set_xlabel("Rabbits")
ax1.set_ylabel("Foxes")

plt.show()
```

The first plot gives us the population of rabbits and foxes as a function of time. Notice how, when the population of the rabbits are initially very small, the population of the foxes begin to drop due to a lack of food. This causes the population of rabbits to increase since they are not being hunted as often. This, in turn, causes an increase in the population of foxes, which then decreases the rabbits. This cycle then repeats itself.

Notice how around when the population of rabbits peak, the population of foxes increases rapidly due to the higher concentration of food.

All of this is logically expected, meaning our graph is accurate.

The second plot gives us a phase diagram for the population of foxes and rabbits. Notice how this forms a limit cycle, meaning a closed trajectory in phase space. This tells us that this cycle is stable.

### PROBLEM 2

**Nonlinear pendulum**

_Refer to Exercise 8.4 in the Newman text._ Building on the results from Example 8.6, calculate the motion of a nonlinear pendulum as follows.

1. Solve the two first-order equations, Eqs. (8.45) and (8.46), using the fourth-order Runge-Kutta method for a pendulum with a 10 cm arm. Use your program to calculate the angle $\theta$ of displacement for several periods of the pendulum when it is released from a standstill at $\theta = 179^\circ$ from the vertical. *(25 pts.)*

2. Make a graph of $\theta$ as a function of time. *(25 pts.)*


```python
# Number 1

import numpy as np
import math

# We initialize values for our constants.

g = 9.81
l = 0.1

# We initialize the given initial values.

theta_0 = 179 * (math.pi/180)
omega_0 = 0

# We define a function f(r) similarly to how we defined our function in Problem 1.
# We store both Equation (8.45) and Equation (8.46) in an array, forming a sort of "vector", if you'd like.

def f(r):
    
    # Again, r will be the initial values for theta and omega.
    
    theta = r[0] # The initial value for theta will be stored in the first entry.
    omega = r[1] # The initial value for omega will be stored in the second entry.
    theta_prime = omega # This is Equation (8.45).
    omega_prime = -(g/l)*math.sin(theta) # This is Equation (8.46)
    f = np.array([theta_prime, omega_prime])
    return f

# We initialize some constants.

t_start = 0 # We will start our graph at 0 seconds.
t_end = 30 # We will end our graph at 30 seconds.
N = 1000 # We will have 1000 steps in between 0 and 30 seconds.

# We encode our initial conditions.

r_0 = np.array([theta_0, omega_0], float) # r_0 gives the initial conditions for theta and omega.

# Like the question asks, we call our previously defined function again. This time, we use the parameters for the pendulum problem.

results_pend = runge_kutta_4(f, r_0, t_start, t_end, N)

# We subset the results of our function to obtain theta and time.

theta = results_pend[0]
t_pend = results_pend[2]
```

```python
# Number 2

# We graph angular displacement against time, like the question asks.

fig, ax = plt.subplots()
plot = ax.plot(t_pend, theta)

# We label our axes.

ax.set_title('Angular Displacement of Nonlinear Pendulum')
ax.set_xlabel('Time t (s)')
ax.set_ylabel('Angular Displacement θ (rad)')

# We show the plot.

plt.show()
```
