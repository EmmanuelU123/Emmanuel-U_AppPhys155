**The Schrodinger equation and the Crank-Nicolson method**

_Refer to Exercise 9.8 in the Newman text._ In this problem, you will use the Crank-Nicolson method to solve the full time-dependent Schrodinger equation and hence develop a picture of how a wavefunction evolves over time.

Consider an electron (mass $M = 9.109 \times 10^{-31}$ kg) in a box of length $L = 10^{-8}$ m. Suppose that at time $t = 0$ the wavefunction of the electron has the form

$$ \psi(x,0) = \exp\left[-\frac{(x-x_0)^2}{2\sigma^2}\right]e^{i\kappa x},$$
where $x_0 = \frac{L}{2}$, $\sigma = 1 \times 10^{-10}$ m, $\kappa = 5 \times 10^{10} {\rm m}^{-1}$,  and $\psi = 0$ on the walls at $x = 0$ and $x = L$.

1. Perform a single step of the Crank-Nicolson method for this electron, calculating the vector $\psi(t)$ of values of the wavefunction, given the initial wavefunction above and using $N = 1000$ spatial slices with $a = L/N$. Your program will have to perform the following steps. First, given the vector $\psi(0)$ at $t = 0$, you will have to multiply by the matrix $\bf{B}$ to get a vector $\bf{v} = \bf{B}\psi$. Because of the tridiagonal form of $\bf{B}$, this is fairly simple. The $i$th component of $\bf{v}$ is given by
$$ v_i = b_1\psi_i + b_2(\psi_{i+1} + \psi_{i-1}).$$

   You will also have to choose a value for the time-step $h$. A reasonable choice is $h = 10^{-18}$ s. *(30 pts.)*

2. Second you will have to solve the linear system ${\bf Ax}= {\bf v}$ for $\bf{x}$, which gives you the new value of $\psi$. You could do this using a standard linear equation solver like the function $\tt solve$ in numpy's $\tt linalg$. *(20 pts.)*

3. Once you have the code in place to perform a single step of the calculation, extend your program to perform repeated steps and hence solve for $\psi$ at a sequence of times a separation $h$ apart. Note that the matrix $\bf A$ is independent of time, so it doesn't change from one step to another. You can set up the matrix just once and then keep on reusing it for every step. *(30 pts.)*

4. Make an animation of the solution by displaying the real part of the wavefunction at each time-step. You can use the function rate from the package visual to ensure a smooth frame-rate for your animation-- see Section 3.5 on page 117 of the Newman text.

   Run your animation for a while and describe what you see. Write a few sentences explaining in physics terms what is going on in the system. *(20 pts.)*

```python
######################################################################
#
# Function to solve a banded system of linear equations using
# Gaussian elimination and backsubstitution
#
# x = banded(A,v,up,down)
#
# This function returns the vector solution x of the equation A.x = v,
# where v is an array representing a vector of N elements, either real
# or complex, and A is an N by N banded matrix with "up" nonzero
# elements above the diagonal and "down" nonzero elements below the
# diagonal.  The matrix is specified as a two-dimensional array of
# (1+up+down) by N elements with the diagonals of the original matrix
# along its rows, thus:
#
#   (  -   -  A02 A13 A24 ...
#   (  -  A01 A12 A23 A34 ...
#   ( A00 A11 A22 A33 A44 ...
#   ( A10 A21 A32 A43 A54 ...
#   ( A20 A31 A42 A53 A64 ...
#
# Elements represented by dashes are ignored -- it doesn't matter what
# these elements contain.  The size of the system is taken from the
# size of the vector v.  If the matrix A is larger than NxN then the
# extras are ignored.  If it is smaller, the program will produce an
# error.
#
# The function is compatible with version 2 and version 3 of Python.
#
# Written by Mark Newman <mejn@umich.edu>, September 4, 2011
# You may use, share, or modify this file freely
#
######################################################################

from numpy import copy

def banded(Aa,va,up,down):

    # Copy the inputs and determine the size of the system
    A = copy(Aa)
    v = copy(va)
    N = len(v)

    # Gaussian elimination
    for m in range(N):

        # Normalization factor
        div = A[up,m]

        # Update the vector first
        v[m] /= div
        for k in range(1,down+1):
            if m+k<N:
                v[m+k] -= A[up+k,m]*v[m]

        # Now normalize the pivot row of A and subtract from lower ones
        for i in range(up):
            j = m + up - i
            if j<N:
                A[i,j] /= div
                for k in range(1,down+1):
                    A[i+k,j] -= A[up+k,m]*A[i,j]

    # Backsubstitution
    for m in range(N-2,-1,-1):
        for i in range(up):
            j = m + up - i
            if j<N:
                v[m] -= A[i,j]*v[j]

    return v
```

```python
import numpy as np
import matplotlib.pyplot as plt

M = 9.109e-31 # Mass of the Electron
L = 1e-8 # Length of the box

# Below are the constants that define the Gaussian function found in our wave function.

x_0 = L/2
sigma = 1e-10
kappa = 5e10

N = 1000 # Number of spatial slices
a = L/N # Length of each discrete spacial step

# We define a function for the initial wave function at t = 0 as given by the problem

def psi_0(x):
  return np.exp(-1.*(x-x_0)**2/(2*sigma*sigma))*np.exp(1j*kappa*x)

# We define an array space_step that contains our discrete points in space. This will be used as the x values for our graph.

space_step = np.linspace(0, L, N+1) # The space_step array goes from 0 to the length of the box L in N+1 discrete steps.

# We create an array psi_initial that contains the values of the initial wave function for each discrete spacial step.

psi_initial = np.array(list(map(psi_0, space_step)), complex)

# We plot the initial wave function

fig, ax = plt.subplots()
plot = ax.plot(space_step, psi_initial)
ax.set_xlabel('Distance (m)')
ax.set_ylabel('Psi')
plt.title("Initial Wave Function")

# We plot the initial wave function's absolute value squared.
# As we know from our QM classes, the wave function's absolute value squared gives the probability density function of the particle's position.

fig, ax1 = plt.subplots()
plot1 = ax1.plot(space_step, abs(psi_initial)**2)
ax1.set_xlabel('Distance (m)')
ax1.set_ylabel('|Psi|^2')
plt.title("Initial Wave Function")
```

At t=0, the initial wave function is very tightly packed in the center.

```python
h = 1e-15 # h is the duration of each time step. The problem suggests 1e-18 but this made computation times too long.
hbar = 1.0546e-36 # Planck's constant divided by 2pi.

# The following are the constants in the entires of the matrix A. This is from Newman's textbook.

a1 = 1 + h*(1j*hbar)/(2*M*a**2)
a2 = -h*(1j*hbar)/(4*M*a**2)

# The following are the constants in the entries of the matrix B. This is also from Newman's textbook.

b1 = 1 - h*(1j*hbar)/(2*M*a**2)
b2 = h*(1j*hbar)/(4*M*a**2)

# We define the matrix A using the syntax required for the banded function.

A = np.zeros((3,N+1),complex)

A[0,:] = a2
A[1,:] = a1
A[2:,] = a2
```

```python
# We define a function that evolves the initial wave function for a given amount of time steps.

psi = np.array(list(map(psi_0, space_step)), complex)

def psi_evolved(steps):
  
  # We construct a new array psi and set it equal to psi_initial.
  
  psi = np.array(list(map(psi_0, space_step)), complex)
  
  # These are the boundary conditions
  
  psi[[0,N]]=0
  
  # We construct a for loop to repeatedly evolve our state psi until the required number of steps is reached.
  
  for i in range(steps):
    v = b1*psi[1:N] + b2*(psi[2:N+1] + psi[0:N-1]) # This gives us the current value of the matrix 
    psi[1:N] = banded(A,v,1,1) # Next, we must solve the linear system Ax = v for x. We accomplish this using the banded algorithm.
                               # We store the result back into psi, which will be iterated again until the required number of steps is reached.
  return psi

# To test the function, we evaluate psi after 800 steps.

steps = 20
psi_20 = np.real(psi_evolved(steps))

# We plot the wave function after 800 steps.

fig, ax = plt.subplots()
plot = ax.plot(space_step, psi_20)
ax.set_xlabel('Distance (m)')
ax.set_ylabel('Psi')
plt.title(f"Time-Dependent Wavefunction {steps} iterations")
```

Looking at the graph above, we note that since the amount of time steps is still small (20), it is still reasonably localized near its origin position. However, looking at it closely, it has already begun spreading out a bit and travelling towards the right side of the box. It is no longer centered in the middle.

Let us see what happens later on in time.

```python
psi_80 = np.real(psi_evolved(80))

fig, ax = plt.subplots()
plot = ax.plot(space_step, psi_80)
ax.set_xlabel('Distance (m)')
ax.set_ylabel('Psi')
plt.title('Time-Dependent Wavefunction (' + str(80) + ") iterations")
```

Notice how after 2800 iterations, the wavefunction has moved significantly to the right. The spread is now also much wider than it was initially, indicating a higher uncertainty in position.

```python
psi_140 = np.real(psi_evolved(140))

fig, ax = plt.subplots()
plot = ax.plot(space_step, psi_140)
ax.set_xlabel('Distance (m)')
ax.set_ylabel('Psi')
plt.title('Time-Dependent Wavefunction (' + str(140) + ") iterations")
```

After 140 iterations, the electron's wave function has begun to reflect off of the right wall and interfere with itself. Interesting!

```python
psi_220 = np.real(psi_evolved(220))

fig, ax = plt.subplots()
plot = ax.plot(space_step, psi_220)
ax.set_xlabel('Distance (m)')
ax.set_ylabel('Psi')
plt.title('Time-Dependent Wavefunction (' + str(220) + ") iterations")
```

Finally, after 220 iterations, the electron's wave function has almost completely bounced off of the right wall and is now moving towards the left wall, where it will reflect again. Let us now animate this wave function.

```python
# We import the relevant packages

from matplotlib import animation

# We initialize a figure with our desired formatting

fig = plt.figure()
ax = plt.axes(xlim=(0, L), ylim=(-1.5, 1.5))
line, = ax.plot([], [], lw=5)
ax.set_xlabel('Distance (m)')
ax.set_ylabel('Psi')
plt.title("Time-Dependent Wavefunction")
line, = ax.step([], [])

# We re-initialize an array for psi since the old one has already been time-evolved

psi = psi_initial
psi[0] = psi[N - 1] = 0

# We initialize an init function that we will use for our animation

def init():
    line.set_data([], [])
    return line,

# We define the animate function, this is the most crucial part of coding the animation

def animate(i):
    
    # We initialize the x and y values for our graph
    
    x = np.linspace(0, L, N+1)
    y = psi_evolved(i)
    line.set_data(x, y)
    line.set_color('red')
    return line,

# We call the animation function on animate(i) to finalize our animation

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=2000, interval=100, blit=True)
anim.save("test.gif", fps = 24)
```

In the animation, we see the wavefunction bounce off the right wall and begin interfering with itself. Notice how the tail of the wave packet is slower than the "front" of the wave packet. This is because even though the wave packet is fairly localized in position space, we can expand it as a sum of infinitely many plane waves in momentum space by taking the Fourier Transform. The plane waves with lower k-value contribute towards the slow moving tail, while the ones with higher k-value contribute towards the fast moving "front".

Notice how due to the difference in speeds, eventually the wave packet just becomes distributed over the whole box, to the point where it no longer even resembles a wave packet. Cool!
